```python
#Exercice 2.1
users = []
num_users = int(input("Enter number of users: "))
for i in range(num_users):
    user = []
    num_activities = int(input(f"User {i}: How many activitites: "))
    for j in range(num_activities):
        user.append(input(f"User {i}: Enter an activity: "))
    
    users.append(user)
print(users)
```

```python
#Store values for later
backup_users = users.copy()
```

```python
#Restore values
users = backup_users.copy()
```

```python
#Exercice 2.2
search_activity = input("Enter activity to search for: ")
total = 0
for user in users:
    for activity in user:
        if search_activity in activity:
            total += 1
            continue; # If the activity is detected multiple times per user, we only want to increment the total once
print(f"{total} people like {search_activity}")
```

```python
#Exercice 2.3
index1 = int(input("Enter user 1: "))
index2 = int(input("Enter user 2: "))

user1 = users[index1]
user2 = users[index2]

mutual_activities = 0
for activity in user1:
    if activity in user2: 
        mutual_activities += 1
print(f"Users {index1} and {index2} have {mutual_activities} mutual activities")
```

```python
#Exercice 2.4
index = int(input("Enter user index: "))
user = users[index]
print(user)
remove_activity = str(input("Enter activity to remove"))
user.remove(remove_activity)
print(user)

```

```python
#Exercice 5.2
from random import randint
students = []
for i in range(10):
    c_project = {
        "demonstration": randint(10, 20),
        "code"         : randint(10, 20),
        "presentation" : randint(10, 20),
        "documentation": randint(10, 20),
    }
    python_project = {
        "demonstration": randint(15, 20),
        "code"         : randint(15, 20),
        "presentation" : randint(15, 20),
        "documentation": randint(15, 20),
    }
    student = {
        "c_project"     : c_project,
        "python_project": python_project
    }
    students.append(student)

```

```python
#Exercise 5.3
for i, student in enumerate(students):
    c_project = student["c_project"]
    python_project = student["python_project"]
    c_grade = (c_project["demonstration"] + c_project["code"] + c_project["presentation"] + c_project["documentation"]) * .25
    python_grade = (python_project["demonstration"] + python_project["code"] + python_project["presentation"] + python_project["documentation"]) * .25

    print(f"Student {i} C: {c_grade}, Python: {python_grade}")
```

```python
#Exercise 5.4
c_average = 0
python_average = 0
for student in students:
    c_project = student["c_project"]
    python_project = student["python_project"]
    c_grade = (c_project["demonstration"] + c_project["code"] + c_project["presentation"] + c_project["documentation"]) * .25
    python_grade = (python_project["demonstration"] + python_project["code"] + python_project["presentation"] + python_project["documentation"]) * .25

    c_average += c_grade / len(students)
    python_average += python_grade / len(students)

print(f"C average: {c_average}, Python average: {python_average}")
```

```python
#Exercise 5.5
scale = python_average / c_average
```

```python
#Exercise 5.6
for student in students:
    c_project = student["c_project"]
    python_project = student["python_project"]
    c_grade = min(scale * (c_project["demonstration"] + c_project["code"] + c_project["presentation"] + c_project["documentation"]) * .25, 20)
    python_grade = (python_project["demonstration"] + python_project["code"] + python_project["presentation"] + python_project["documentation"]) * .25

    student["c_grade"] = c_grade
    student["python_grade"] = python_grade

for i, student in enumerate(students):
    print(f"Student {i} C: {student['c_grade']}, Python: {student['python_grade']}")
```
